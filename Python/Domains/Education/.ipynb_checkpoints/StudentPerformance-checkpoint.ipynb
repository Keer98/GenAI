```python
import pandas as pd
import numpy as np
import kagglehub
import os
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
path = kagglehub.dataset_download("devansodariya/student-performance-data")

print("Path to dataset files:", path)
print("Path to dataset files:", path)
csv_file = None
for file in os.listdir(path):
    if file.endswith('.csv'):
        csv_file = os.path.join(path, file)
        break

if csv_file is None:
    print("No CSV file found in the directory.")
else:
    print(f"Found CSV file: {csv_file}")
```

```python
df = pd.read_csv(csv_file, delimiter=',')
df.head()
```

```python
rows, columns = df.shape

print(f"Number of rows: {rows}")
print(f"Number of columns: {columns}")

print(df.info())
```

```python
for column in df.columns:
    print(f"\nUnique values and counts for column '{column}':")
    print(df[column].value_counts())
```

```python
df.describe()
```

```python
df['Grade'] = df[['G1', 'G2', 'G3']].mean(axis=1)
df.head()
```

```python
df.drop(['G1', 'G2', 'G3'],axis=1,inplace=True)
```

```python
X_df=df.drop('Grade',axis=1)
Y_df=df['Grade']
```

```python
num_plots = len(X_df)
plots_per_row = 5
num_rows = (num_plots + plots_per_row - 1) // plots_per_row  # Calculate number of rows needed

plt.figure(figsize=(15, 5 * num_rows))  # Adjust height based on number of rows

# Loop through each column and create a bar plot
for i, column in enumerate(X_df):
    plt.subplot(num_rows, plots_per_row, i + 1)  # Create subplots
    sns.barplot(x=column, y='Grade', data=df, errorbar=None)  # Use mean for aggregation
    plt.title(f'Average Grade by {column}')
    plt.xlabel(column)
    plt.ylabel('Average Grade')

    plt.xticks(rotation=45)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8)) 
numerical_df = df.select_dtypes(exclude=['object'])
sns.heatmap(numerical_df.corr(), annot=True, fmt=".2f")
```

#Converting boolean to numeric
[schoolsup,famsup,paid,activities,nursery,internet,romantic]

```python
boolean_cols = ['schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'internet', 'romantic']

X_df[boolean_cols] = X_df[boolean_cols].replace({'yes': 1, 'no': 0})
```

#Ordinal-> Label Encoding
#Nominal[school,sex,famsize,Pstatus,Mjob,Fjob,reason,guardian] -> One-Hot Encoding

```python
# Apply One-Hot Encoding for Nominal Variables
nominal_cols = ['school', 'address','sex', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 'reason', 'guardian','higher']
df_encoded = pd.get_dummies(X_df, columns=nominal_cols, drop_first=True)
df_encoded.info()
df_encoded = df_encoded.astype(int)
df_encoded.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_encoded, Y_df, test_size=0.2, random_state=42)
# Apply Standardization to the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

# Display the predictions and evaluation metrics
results_df = pd.DataFrame({'Actual': y_test, 'Predicted': predictions})
print(results_df)

print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```
